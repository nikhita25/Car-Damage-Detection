```python
IMG_DIR = '/Users/rohitgulve/Documents/Applied_Machine_Learning/data3a/training/01-minor'


for img in os.listdir(IMG_DIR):
    img_array = cv2.imread(os.path.join(IMG_DIR,img), cv2.IMREAD_GRAYSCALE)
    
    img_pil = Image.fromarray(img_array)
    img_100x100 = np.array(img_pil.resize((100, 100), Image.ANTIALIAS))

    img_array = (img_100x100.flatten())

    img_array  = img_array.reshape(-1,1).T

    with open('/Users/rohitgulve/Documents/Applied_Machine_Learning/data3a/training_data_minor_01.csv', 'ab') as f:

        np.savetxt(f, img_array, delimiter=",")
```
